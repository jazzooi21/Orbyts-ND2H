## Read from a file
Function for reading from a file and writing to a file.

```python
def readFile(filename, columnNames=None):
    if columnNames == None:
        dataFrame = pd.read_csv(filename, delim_whitespace=True)
    else:
        dataFrame = pd.read_csv(filename, delim_whitespace=True, names=columnNames)
    return dataFrame

def writeToFile(dataFrame, fileName):
    dataFrame = dataFrame.to_string(index=False)
    with open(fileName, "w+") as fileToWriteTo:
        fileToWriteTo.write(dataFrame)
```

The `delim_whitspace=True` argument is applied for when columns in the file are seperated by whitespace instead of columns

### Apply function to row of a DataFrame
Function for applying a function to each row of a data frame.

```python
def applyFunction(function, dataFrame):
    dataFrame = dataFrame.apply(lambda x:function(x), result_type="expand", axis=1)
    return dataFrame
```

```python
def generateTags(row):
    row["Tag'"] = (str(row["nu1'"]) + "-" + str(row["nu2'"]) + "-" + str(row["nu3a'"]) 
                   + "-" + str(row["nu3b'"]) + "-" + str(row["nu4a'"]) + "-" + str(row["nu4b'"])
                   + "-" + str(row["J'"]) + "-" + str(row["Ka'"]) + "-" + str(row["Kc'"])  
                   + "-" + str(row["inv'"]))
    row["Tag\""] = (str(row["nu1\""]) + "-" + str(row["nu2\""]) + "-" + str(row["nu3a\""]) 
                   + "-" + str(row["nu3b\""]) + "-" + str(row["nu4a\""]) + "-" + str(row["nu4b\""])
                   + "-" + str(row["J\""]) + "-" + str(row["Ka\""]) + "-" + str(row["Kc\""])  
                   + "-" + str(row["inv\""]))
    return row

transitions = applyFunction(generateTags, transitions)
transitions
```

```python
transitions = transitions.merge(empiricalEnergies[["Tag\"", "E/CM-1"]])
# Rename column and discard
transitions["E\""] = transitions["E/CM-1"]
transitions = transitions.drop("E/CM-1", axis=1)
transitions["E'"] = transitions["nu"] + transitions["E\""]
```

### Group apply function
Function to be applied to dataframes grouped by a certain column

```python
def groupApplyFuction(function, groupByColumn, dataFrame):
    groupedDataFrame = dataFrame.groupby([groupByColumn])
    dataFrame = groupedDataFrame.apply(lambda x:function(x))
    return dataFrame
```

```python
def applyCombinationDifferences(dataFrame):
    dataFrame["Eavg'"] = dataFrame["E'"].mean()
    dataFrame["Problem"] = abs(dataFrame["Eavg'"] - dataFrame["E'"].mean()) > 0.05 
    dataFrame["Return"] = dataFrame["Problem"].any()
    return dataFrame

transitions = groupApplyFuction(applyCombinationDifferences, "Tag'", transitions)
transitions[["Source", "Tag'", "Tag\"", "E'", "unc1","Eavg'", "Problem","E\""]]
# problematicTransitions = transitions[transitions["Return"]]
# problematicTransitions[["Source", "Tag'", "Tag\"", "E'", "unc1","Eavg'", "Problem","E\""]]
```
